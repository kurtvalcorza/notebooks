### 5. Audio Format Conversion (skip if already working with supported audio files)

This step converts audio files from M4A to MP3 format. Converting to MP3 is often necessary to ensure compatibility with various processing tools or to reduce file size while maintaining acceptable quality before transcription.

The FFmpeg command uses the following flags:
- `-i`: Specifies the input file path.
- `-vn`: Disables video recording (ensures only audio is processed).
- `-ar`: Sets the audio sampling rate (e.g., 44100 Hz).
- `-ac`: Sets the number of audio channels (e.g., 2 for stereo).
- `-b:a`: Sets the audio bitrate (e.g., 192k) to control output quality and size.

````python
input_file = '/content/drive/MyDrive/Voice 251201_131937.m4a' #@param {type: "string"}
output_file = '/content/drive/MyDrive/Voice 251201_131937.mp3' #@param {type: "string"}

!ffmpeg -i "{input_file}" -vn -ar 44100 -ac 2 -b:a 192k "{output_file}"
````

# 6. Batch Transcription

Python script to automate the transcription of multiple files located in a specific Google Drive directory.

The script performs the following logic:
1. **Iterates** through all files in the defined input directory.
2. **Filters** for files with the `.mp3` extension.
3. **Executes** the Whisper CLI command for each identified file.

The Whisper command utilizes the `large-v3` model and saves the generated output files (such as text, VTT, and SRT formats) back to the specified output directory.

````python
import os

# Define the input directory containing audio files
input_dir = '/content/drive/MyDrive/'  #@param {type: "string"}

# Iterate over audio files in the input directory
for audio_file in os.listdir(input_dir):
    if audio_file.endswith('.mp3'):  # Adjust the file extension as needed
        audio_path = os.path.join(input_dir, audio_file)

        # Transcribe the audio file
        !whisper "{audio_path}" --model large-v3 --output_dir "{input_dir}"
````
